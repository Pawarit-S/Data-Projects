```python
import pandas as pd
import seaborn as sns
import numpy as np
```

read csv file and inspect all columns

```python
nobel = pd.read_csv('data/nobel.csv')
print(nobel.columns)
```

1.count values on sex,country
2.index max count by gender,country

```python
top_gender = nobel['sex'].value_counts().idxmax()
top_country = nobel['birth_country'].value_counts().idxmax()
print(top_gender)
print(top_country)
```

1.Add decade columb to DataFrame
2.count total winners each decade
3.count total USA winners each decade
4.index max total USA winners each decade

```python
nobel['decade'] = (nobel['year'] // 10) * 10
total_winners_by_decade = nobel.groupby('decade').size()
us_winners_by_decade = nobel[nobel['birth_country'] == 'United States of America'].groupby('decade').size()
max_decade_usa = (us_winners_by_decade/total_winners_by_decade).idxmax()
```

1.count total winner by decade and category
2.subset Female gender and count total Female winner by decade and category
3.index max()
4.dictionary

```python
total_winners_by_decade_cat = nobel.groupby(['decade','category']).size()
female_winners_by_decade_cat = nobel[nobel['sex']=='Female'].groupby(['decade','category']).size()
proportion_female_by_decade_cat = female_winners_by_decade_cat / total_winners_by_decade_cat
highest_proportion_female_by_decade_cat = proportion_female_by_decade_cat.idxmax()
max_female_dict = {highest_proportion_female_by_decade_cat[0]:highest_proportion_female_by_decade_cat[1]}
```

1.subset female winner
2.index first year of female winner
3.show full name and category

```python
female_winners = nobel[nobel['sex'] == 'Female']
first_woman_index = female_winners['year'].idxmin()
first_woman_name = female_winners.loc[first_woman_index,'full_name']
first_woman_category = female_winners.loc[first_woman_index,'category']
print(first_woman_name)
print(first_woman_category)
```

1.count total win each nobel winners
2.subset nobel winner more than 1 time (2 or more)
3.create list repeat nobel prize winner

```python
nobel_winner_counts = nobel.groupby('full_name').size()
nobel_winner_repeat = nobel_winner_counts[nobel_winner_counts >= 2]
repeat_list = nobel_winner_repeat.index.to_list()
print(repeat_list)
```
