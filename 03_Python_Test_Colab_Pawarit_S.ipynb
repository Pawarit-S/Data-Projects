Import panda and read csv each table

```python
import pandas as pd
pricing_project_dataset = pd.read_csv("/content/drive/MyDrive/pricing_project_dataset.csv")
platform_number = pd.read_csv("/content/drive/MyDrive/platform_number.csv")
exchange_rate = pd.read_csv("/content/drive/MyDrive/exchange_rate.csv")
```

Calculate the Order Coverge

```python
pricing_platform = pd.merge(pricing_project_dataset, platform_number, left_on='grass_region', right_on='region')
pricing_platform['platform order'] = pd.to_numeric(pricing_platform['platform order'], errors='coerce')
pricing_platform['Order Coverage'] = pricing_platform['shopee_order'] / pricing_platform['platform order']
pricing_platform.head()
```

Calculate the net competitiveness metric

```python
pricing_platform['net competitiveness diff'] = pricing_platform[(pricing_platform['shopee_model_competitiveness_status']=='Shopee < CPT')].count() - pricing_platform[(pricing_platform['shopee_model_competitiveness_status']=='Shopee > CPT')].count()
pricing_platform['net competitiveness'] = pricing_platform['net competitiveness diff'] / pricing_platform['shopee_model_competitiveness_status'].count()
```

```python
summary = pricing_platform.groupby('grass_region').agg({'Order Coverage': 'sum', 'net competitiveness': 'sum','shopee_order':'sum'})
summary.head()
```
